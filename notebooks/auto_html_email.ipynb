```python
import pandas as pd

# Load performance data
df = pd.read_csv("../dashboards/fact_performance.csv")

# Top 5 funds by Sharpe Ratio
top_funds = df.sort_values(
    by="sharpe_ratio",
    ascending=False
).head(5)

# Create HTML report
html_content = f"""
<html>
<head>
<title>Weekly Mutual Fund Report</title>
</head>
<body>

<h1>Bluestock Weekly Mutual Fund Summary</h1>

<h2>Top 5 Funds by Sharpe Ratio</h2>

{top_funds[['amfi_code',
            'return_5yr_pct',
            'sharpe_ratio',
            'risk_grade']].to_html(index=False)}

<p>
Generated automatically using Python.
</p>

</body>
</html>
"""

# Save HTML file
with open("weekly_report.html", "w", encoding="utf-8") as f:
    f.write(html_content)

print("HTML Report Generated Successfully!")
```
